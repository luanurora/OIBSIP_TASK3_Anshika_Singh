```python
# Importing libraries
import pandas as pd
```

```python
# Loading the dataset
data = pd.read_csv("C:/Users/Anshika/OneDrive/Desktop/AB_NYC_2019.csv")

# Displaying to verify
data.head()
```

```python
# Handling the missing values columns
data['name'].fillna('NA', inplace=True)
data['host_name'].fillna('NA', inplace=True)
data['last_review'].fillna('1970-01-01', inplace=True)
data['reviews_per_month'].fillna(0, inplace=True)

# Displaying to verify 
data.head()
```

```python
# Removing duplicate rows 
data.drop_duplicates(inplace=True)

# Displaying the number of rows to verify 
data.shape
```

```python
# Standardizing 'name' and 'host_name' to have consistent capitalization
data['name'] = data['name'].str.title()
data['host_name'] = data['host_name'].str.title()

# Converting 'last_review' to datetime format
data['last_review'] = pd.to_datetime(data['last_review'])
 
data.head()
```

```python
# Removing listings with price above the 99th percentile
price_threshold = data['price'].quantile(0.99)
data = data[data['price'] < price_threshold]

data.shape
```

```python
# Ensuring 'latitude' and 'longitude' are within valid ranges
data = data[(data['latitude'].between(-90, 90)) & (data['longitude'].between(-180, 180))]

# Ensuring 'price' and 'minimum nights' have reasonable values
data = data[(data['price'] > 0) & (data['minimum_nights'] > 0)]

data.shape
```

```python
# Saveing the cleaned data
cleaned_data = "C:/Users/Anshika/OneDrive/Desktop/AB_NYC_2019_cleaned.csv"
data.to_csv(cleaned_data, index=False)

# Displaying the path of the cleaned file
cleaned_data
```
